```
from urllib import request
from bs4 import BeautifulSoup

content = request.urlopen("https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnl=109")
soup = BeautifulSoup(content, 'html.parser')

print( soup.select("data"))
```

```
for data in soup.select("data") :
  print("time : ", data.select_one("tmef").string)
  print("weather : ", data.select_one("wf").string)
  print("-" * 10)
```
